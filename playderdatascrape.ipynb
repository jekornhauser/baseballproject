```python
from splinter import Browser
import time
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
import pandas as pd
```

```python
db = pd.read_csv("player_urls.csv")
player_urls = db["Player_URL"]
```

```python
name = []
position = []
year = []
age = []
team = []
league = []
games_played = []
at_bats = []
runs = []
hits = []
doubles = []
triples = []
homeruns = []
rbis = []
stolen_bases = []
strikeouts = []
batting_average = []
on_base_percentage = []
slugging_percent = []
ops = []
urls = []
dob = []
```

```python
for n,i in enumerate(player_urls, start=1):
    URL = "https://www.baseball-reference.com/" + i
    browser.visit(URL)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    details = soup.find(itemtype="https://schema.org/Person")
    header = details.find("h1")
    player_name = header.text
    player_name = player_name.strip('\n    ')
    pos = details.find("p")
    posit = pos.text
    posit = posit.strip('Position:\n')
    posit = posit.strip('\n  ')
    dobdetails = soup.find(id="necro-birth")
    birthdate = dobdetails.text
    birthdate = birthdate.strip('\n')
    r = soup.find(id='batting_standard')
    table = r.find("tbody")
    rows = table.find_all("tr")
    for row in rows:
        try:
            year_played = row.find("th")
            data = row.find_all('td')
            hits.append(data[7].text)
            name.append(player_name)
            position.append(posit)
            year.append(year_played.text)
            age.append(data[0].text)
            team.append(data[1].text)
            league.append(data[2].text)
            games_played.append(data[3].text)
            at_bats.append(data[5].text)
            runs.append(data[6].text)
            doubles.append(data[8].text)
            triples.append(data[9].text)
            homeruns.append(data[10].text)
            rbis.append(data[11].text)
            stolen_bases.append(data[12].text)
            strikeouts.append(data[15].text)
            batting_average.append(data[16].text)
            on_base_percentage.append(data[17].text)
            slugging_percent.append(data[18].text)
            ops.append(data[19].text)
            urls.append(i)
            dob.append(birthdate)
        except IndexError:
            continue
    print("The statistics of player " + player_name + " have been added.")
    print("You have compeleted adding statistics of " + str(n) + " players out of a total of 2367." )
    print("-------------------------------------")
    time.sleep(3)
```
